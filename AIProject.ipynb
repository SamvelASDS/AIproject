```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Generating user data from https://randomuser.me/**

```python
URL = 'https://randomuser.me/api/?results=5000&format=csv'
#Such as I can't generate more than 5000 users.
data1 = pd.read_csv(URL)
data2 = pd.read_csv(URL)
finalData = pd.concat([data1, data2], ignore_index=True) #Generated 10.000 users
```

**Pandas dataframe from generated data.**

```python
pd.DataFrame(finalData) #Or we can just type "finalData" ,the result is same.
```

**Performing EDA on that data.**

```python
finalData.info()
finalData.describe()
```

```python
df = finalData
```

**Gender visualizations**

```python
gender = df['gender'].value_counts()
gender
```

```python
plt.bar(df['gender'].unique(), gender.array, tick_label = df['gender'].unique(),
        width = 0.7, color = ['green', 'red'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Quantity for gender')
```

```python
plt.pie(gender, labels = df['gender'].unique(), autopct='%1.1f%%', radius=2, shadow=True)
```

**Name visualizations**

**Age visualizations**

```python
df['dob.age'].describe()
```

```python
age = df['dob.age'].value_counts()
age
```

```python
ageSorted = np.sort(df['dob.age'])
plt.hist(ageSorted,width = 0.01, color = 'blue', edgecolor='black')
plt.hist(ageSorted)
plt.xlabel('Age')
plt.ylabel('Count')
```

**Country visualizations**

```python
country = df['location.country'].value_counts()
country
```

```python
plt.pie(country, labels = df['location.country'].unique(), autopct='%1.1f%%', radius=5, shadow=True)
plt.legend(bbox_to_anchor=(4,2), )
```

```python
ax1 = plt.axes([0, 0, 3, 2])
plt.bar(df['location.country'].unique(), country.array, tick_label = df['location.country'].unique(),
        width = 0.8, color = ['brown', 'blue'])
plt.xlabel('Countries')
plt.ylabel('Number of users')
plt.title('Distribution of users by countries ')
```
